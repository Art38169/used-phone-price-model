## Binary columns

```python
binary_cols = [
    "has_5g",
    "screen_cracked",
    "body_damage",
    "repair_history",
    "water_damage",
    "box_available",
    "charger_available",
]
```

```python
import pandas as pd

summary = []

for col in binary_cols:
    avg = df.groupby(col)["resale_price"].mean()

    summary.append({
        "Feature": col,
        "Mean (0)": avg.loc[0],
        "Mean (1)": avg.loc[1],
        "% Change": ((avg.loc[1] - avg.loc[0]) / avg.loc[0]) * 100,
    })

summary_df = (
    pd.DataFrame(summary)
    .sort_values("% Change")
)

summary_df
```
